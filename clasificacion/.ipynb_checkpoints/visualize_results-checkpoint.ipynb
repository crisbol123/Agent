# SLM Classification Evaluation — Results Visualization
Compara los dos runs de evaluación sobre 800 muestras (100/categoría, 8 categorías, 5 modelos).

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted', font_scale=1.1)

FILES = {
    'Run 1 (203457)': 'classification_results_20260314_203457.json',
    'Run 2 (213410)': 'classification_results_20260314_213410.json',
}

runs = {}
for label, path in FILES.items():
    with open(path) as f:
        runs[label] = json.load(f)

print('Runs cargados:', list(runs.keys()))
```

## 1 — Resumen de métricas globales por modelo

```python
for run_label, data in runs.items():
    rows = []
    for r in data['results']:
        rows.append({
            'Modelo': r['model_name'],
            'Accuracy': r['accuracy'],
            'Precision': r['precision'],
            'Recall': r['recall'],
            'F1': r['f1_score'],
            'Tiempo/muestra (s)': r['avg_time_per_sample'],
        })
    df = pd.DataFrame(rows).set_index('Modelo')
    print(f'\n=== {run_label} ===')
    display(df.style.format('{:.4f}').background_gradient(cmap='YlGn', subset=['Accuracy','F1']))
```

## 2 — Accuracy y F1 por modelo (barras agrupadas)

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5), sharey=True)

for ax, (run_label, data) in zip(axes, runs.items()):
    models = [r['model_name'].replace('-Instruct','').replace('-Chat','') for r in data['results']]
    acc    = [r['accuracy']  for r in data['results']]
    f1     = [r['f1_score']  for r in data['results']]
    prec   = [r['precision'] for r in data['results']]

    x = np.arange(len(models))
    w = 0.25
    bars_acc  = ax.bar(x - w, acc,  w, label='Accuracy',  color='steelblue')
    bars_f1   = ax.bar(x,     f1,   w, label='F1 macro',  color='coral')
    bars_prec = ax.bar(x + w, prec, w, label='Precision', color='mediumseagreen')

    for bars in [bars_acc, bars_f1, bars_prec]:
        for bar in bars:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                    f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=8)

    ax.set_title(run_label)
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=20, ha='right')
    ax.set_ylim(0, 0.75)
    ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
    ax.legend()

fig.suptitle('Métricas globales por modelo', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('plot_metrics_by_model.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3 — Recall por categoría (heatmap)

```python
import re

def parse_recall_per_category(report_str, categories):
    """Extrae recall por categoría del classification_report string."""
    recalls = {}
    for cat in categories:
        pattern = rf'^\s*{re.escape(cat)}\s+[\d.]+\s+([\d.]+)'
        m = re.search(pattern, report_str, re.MULTILINE)
        recalls[cat] = float(m.group(1)) if m else 0.0
    return recalls

fig, axes = plt.subplots(1, 2, figsize=(18, 5))

for ax, (run_label, data) in zip(axes, runs.items()):
    categories = data['categories']
    models = [r['model_name'].replace('-Instruct','').replace('-Chat','') for r in data['results']]
    matrix = []
    for r in data['results']:
        recalls = parse_recall_per_category(r['classification_report'], categories)
        matrix.append([recalls[c] for c in categories])

    df_heat = pd.DataFrame(matrix, index=models, columns=categories)
    sns.heatmap(df_heat, ax=ax, annot=True, fmt='.2f', cmap='RdYlGn',
                vmin=0, vmax=1, linewidths=0.5, cbar_kws={'label': 'Recall'})
    ax.set_title(f'Recall por categoría — {run_label}')
    ax.set_xlabel('Categoría')
    ax.set_ylabel('Modelo')

plt.tight_layout()
plt.savefig('plot_recall_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4 — Matrices de confusión por modelo

```python
for run_label, data in runs.items():
    categories = data['categories']
    n = len(data['results'])
    fig, axes = plt.subplots(1, n, figsize=(5 * n, 4.5))
    fig.suptitle(f'Matrices de confusión — {run_label}', fontsize=13, fontweight='bold')

    for ax, r in zip(axes, data['results']):
        cm = np.array(r['confusion_matrix'])
        # Normalizar por fila (recall)
        cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
        sns.heatmap(cm_norm, ax=ax, annot=cm, fmt='d', cmap='Blues',
                    xticklabels=categories, yticklabels=categories,
                    vmin=0, vmax=1, cbar=False, linewidths=0.3, annot_kws={'size': 8})
        model_short = r['model_name'].replace('-Instruct','').replace('-Chat','')
        ax.set_title(f'{model_short}\nacc={r["accuracy"]:.3f}', fontsize=10)
        ax.set_xlabel('Predicho')
        ax.set_ylabel('Real')
        ax.tick_params(axis='x', rotation=45)
        ax.tick_params(axis='y', rotation=0)

    plt.tight_layout()
    safe_label = run_label.replace(' ', '_').replace('(','').replace(')','').replace('/','_')
    plt.savefig(f'plot_confusion_{safe_label}.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## 5 — Comparación de accuracy entre runs

```python
run_labels = list(runs.keys())
all_models = [r['model_name'].replace('-Instruct','').replace('-Chat','') for r in runs[run_labels[0]]['results']]

acc_by_run = {
    label: [r['accuracy'] for r in data['results']]
    for label, data in runs.items()
}

x = np.arange(len(all_models))
w = 0.35
fig, ax = plt.subplots(figsize=(10, 5))

colors = ['steelblue', 'coral']
for i, (label, accs) in enumerate(acc_by_run.items()):
    bars = ax.bar(x + (i - 0.5) * w, accs, w, label=label, color=colors[i], alpha=0.85)
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(all_models, rotation=15, ha='right')
ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1))
ax.set_ylim(0, 0.65)
ax.set_title('Accuracy por modelo — comparación entre runs', fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('plot_accuracy_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6 — Velocidad de inferencia (tiempo promedio por muestra)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4), sharey=False)

for ax, (run_label, data) in zip(axes, runs.items()):
    models = [r['model_name'].replace('-Instruct','').replace('-Chat','') for r in data['results']]
    times  = [r['avg_time_per_sample'] for r in data['results']]
    colors_bar = ['tomato' if t > 1.0 else 'steelblue' for t in times]
    bars = ax.barh(models, times, color=colors_bar)
    for bar, t in zip(bars, times):
        ax.text(t + 0.01, bar.get_y() + bar.get_height()/2,
                f'{t:.3f}s', va='center', fontsize=9)
    ax.set_xlabel('Segundos por muestra')
    ax.set_title(f'Velocidad — {run_label}')
    ax.axvline(1.0, color='red', linestyle='--', linewidth=0.8, label='1s threshold')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('plot_inference_speed.png', dpi=150, bbox_inches='tight')
plt.show()
```
