# Visualize Generation Results
Este notebook carga el ultimo archivo generation_results_*.json y genera tablas y graficas para comparar modelos.

```python
from pathlib import Path
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 160)

ROOT = Path('.')
result_files = sorted(ROOT.glob('generation_results_*.json'))
if not result_files:
    raise FileNotFoundError('No se encontraron archivos generation_results_*.json en la raiz del proyecto')

RESULT_PATH = result_files[-1]
print(f'Usando archivo: {RESULT_PATH}')
```

```python
with open(RESULT_PATH, 'r', encoding='utf-8') as f:
    payload = json.load(f)

df = pd.DataFrame(payload['results']).copy()

cols = [
    'model_name', 'params', 'samples_evaluated',
    'rouge1', 'rouge2', 'rougeL', 'bertscore_f1',
    'avg_time_per_sample', 'error_count', 'error_rate'
]
df_view = df[cols].sort_values('bertscore_f1', ascending=False).reset_index(drop=True)
df_view
```

## Ranking por calidad

```python
ranking = df_view[['model_name', 'rouge1', 'rouge2', 'rougeL', 'bertscore_f1']].copy()
ranking
```

```python
plot_df = df_view.sort_values('bertscore_f1', ascending=True)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
metrics = ['rouge1', 'rouge2', 'rougeL', 'bertscore_f1']
titles = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'BERTScore-F1']

for ax, metric, title in zip(axes.flatten(), metrics, titles):
    ax.barh(plot_df['model_name'], plot_df[metric])
    ax.set_title(title)
    ax.set_xlim(0, 1)

plt.tight_layout()
plt.show()
```

## Trade-off calidad vs velocidad

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(df_view['avg_time_per_sample'], df_view['bertscore_f1'], s=90)

for _, row in df_view.iterrows():
    ax.annotate(row['model_name'], (row['avg_time_per_sample'], row['bertscore_f1']), xytext=(6, 4), textcoords='offset points')

ax.set_xlabel('Tiempo promedio por muestra (s)')
ax.set_ylabel('BERTScore-F1')
ax.set_title('Trade-off: calidad semantica vs latencia')
plt.show()
```

## Score combinado (opcional)
Se normaliza calidad y velocidad para obtener un ranking balanceado.

```python
tmp = df_view.copy()

quality = tmp['bertscore_f1']
speed = tmp['avg_time_per_sample']

quality_n = (quality - quality.min()) / (quality.max() - quality.min() + 1e-9)
speed_n = 1 - (speed - speed.min()) / (speed.max() - speed.min() + 1e-9)

tmp['quality_norm'] = quality_n
tmp['speed_norm'] = speed_n
tmp['score_balanced'] = 0.7 * tmp['quality_norm'] + 0.3 * tmp['speed_norm']

tmp[['model_name', 'bertscore_f1', 'avg_time_per_sample', 'score_balanced']].sort_values('score_balanced', ascending=False).reset_index(drop=True)
```

## Exportar resumen

```python
out_csv = Path('generation_results_summary.csv')
df_view.to_csv(out_csv, index=False)
print(f'Resumen guardado en: {out_csv}')
```
